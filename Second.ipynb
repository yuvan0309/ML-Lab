```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
```

```python
calfornia_data = fetch_california_housing(as_frame=True)
df = calfornia_data.frame
```

```python
print(df.corr(numeric_only=True))
```

```python
dataplot = sns.heatmap(df.corr(numeric_only=True), cmap="YlGnBu", annot=True)
plt.show()
```

```python
sns.pairplot(df,height=2,plot_kws={'alpha':0.5,'s':10,'edgecolor':'k'})
```
