```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('/home/yuvan/Documents/ML Lab/Datasets/housing.csv')
print(df)
```

```python
df.head()
```

```python
ifeatures=df.select_dtypes(include=[np.number])
ifeatures
```

```python
for feature in ifeatures :
    plt.figure(figsize=(10,5))
    sns.histplot(df[feature],bins=30,kde=True)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
    
```

```python
for feature in ifeatures :
    plt.figure(figsize=(10,5))
    sns.boxplot(x=df[feature])
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.show()
```

```python
for feature in ifeatures:
    q1=np.percentile(df[feature].dropna(),25)
    q3=np.percentile(df[feature].dropna(),75)
    iqr=(q3-q1)
    
    ubound=q1 - 1.5 * iqr
    lbound=q3 + 1.5 * iqr

    outliers=df[(df[feature] < lbound ) | (df[feature] > ubound )]
    
    print(f" \nFeature of :{feature}")
    print(f"Inter Quartile Range:",iqr)
    print(f"25% of {feature}:",q1)
    print(f"75% of {feature}:",q3)
    print(f"\noutliers of {feature}")
    print(f"Number of outliers : {len(outliers)}")
```
